```python
genderDosage = bxd.groupby("Gender")["Dosage_mg"].agg(["count", "mean"]).reset_index().sort_values("mean", ascending = False)
```

```python
genderDosage
```

```python
genderDosage.plot(x = "Side_Effects", y = "mean",
                                        kind = "pie",
                                        autopct = "%1.1f%%",
                                        shadow = True)

plt.title("Cinsiyetlere Gre Ortalama la Doz Kullanm")
plt.show()
```

```python
conditionImpro = bxd.groupby("Condition")["Improvement_Score"].agg(["count", "mean"]).reset_index().sort_values("mean", ascending = False)
```

```python
conditionImpro
```

```python
conditionImpro.plot(x = "Condition", y = "mean",
                   kind = "bar",
                   color = "#0076CE")

plt.title("Hastalklara Gre yileme Puan Ortalamas")
plt.xlabel("Hastalklar")
plt.ylabel("Hastala Gre yileme Puan Dalm")
plt.legend(["yileme Puan Ortalamas"], fontsize = 6)
plt.show()
```

```python
conditionT = bxd.groupby("Condition")["Treatment_Duration_days"].agg(["count", "mean"]).reset_index().sort_values("mean", ascending = False)
```

```python
conditionT
```

```python
conditionT.plot(x = "Condition", y = "mean",
               kind = "bar",
               color = "#0071c5")

plt.title("Hastalklara Gre Ortalama yileme Sresi (Gn Olarak)")
plt.xlabel("Hastalklar")
plt.ylabel("Ortalama yileme Sresi Dalm")
plt.legend(["Ortalama yileme Sresi"], fontsize = 7)
plt.xticks(rotation = 45)
plt.show()
```

```python
ageSideEff = bxd.groupby("Side_Effects")["Age"].agg(["count", "mean"]).reset_index().sort_values("mean", ascending = False)
```

```python
ageSideEff
```

```python
ageSideEff.plot(x = "Side_Effects", y = "mean",
               kind = "bar",
               color = "#1F75FE")

plt.title("Ya Ortalamasna Gre Yan Etkiler")
plt.xlabel("Yan Etkiler")
plt.ylabel("Yan Etkilere Gre Ya Ortlama Dalm")
plt.legend(["Yan etkilere Gre Ya Ortalamas"], fontsize = 9)
plt.show()
```
